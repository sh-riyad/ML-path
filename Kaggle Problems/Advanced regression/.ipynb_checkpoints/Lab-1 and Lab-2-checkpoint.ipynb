```python
import pandas as pd
import numpy as np


# for showing all columns
pd.set_option("display.max_columns", None)
```

```python
dataset = pd.read_csv("data/train.csv")
print(f"Dataset shape : {dataset.shape}")
```

```python
# First 10 rows
dataset.drop(columns=['Id'],inplace=True)
dataset.head()
```

```python
# Last 10 rows
dataset.tail(10)
```

```python
# information about columns
dataset.info()
```

<h2>Missing value</h2>

```python
# finding columns name that have atleast one missing value
missing_columns = [col for col in dataset.columns if dataset[col].isna().sum()>1]
print(f"{len(missing_columns)} columns have missing values")
```

```python
# Counting the number of missing vlues 
dataset[missing_columns].isna().sum().sort_values(ascending=False)
```

<h4>Removing columns that contail too much missing values</h4>

```python
columns_to_remove = ['PoolQC','MiscFeature','Alley','Fence','MasVnrType','FireplaceQu','LotFrontage']
dataset.drop(columns_to_remove, axis=1,inplace=True)
```

<h3>Missing value Imputation</h3>

```python
# Finding missing value with numerical feature
numerical_missing_coloumns = [column for column in dataset.columns if dataset[column].isna().sum() >0 and dataset[column].dtype != 'O']
numerical_missing_coloumns
```

```python
dataset[numerical_missing_coloumns].head()
```

```python
# Mean imputation for 'MasVnrArea' and Median imputation for 'GarageYrBlt'
dataset['MasVnrArea'].fillna(dataset['MasVnrArea'].mean(), inplace= True)
dataset['GarageYrBlt'].fillna(dataset['GarageYrBlt'].median(), inplace= True)
```

```python
# Finding missing value with categorical feature
categorical_missing_coloumns = [column for column in dataset.columns if dataset[column].isna().sum() >0 and dataset[column].dtype == 'O']
categorical_missing_coloumns
```

```python
dataset[categorical_missing_coloumns].head()
```

```python
# Mode imputation for categorical columns
for col in categorical_missing_coloumns:
    dataset.fillna(dataset[col].mode(),inplace=True)
```

<h2>Scaling Numerical Features</h2>

```python
# Finding Numerical columns
numerical_columns = [column for column in dataset.columns if dataset[column].dtypes != 'O']
print(f"Total number of numerical columns {len(numerical_columns)}")
```

```python
# Scaling Numerical columns using MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
for column in numerical_columns:
    dataset[column] = scaler.fit_transform(dataset[[column]])
```
