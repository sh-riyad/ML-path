```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Activation, Flatten
from keras.models import Sequential

%matplotlib inline
```

```python
# importing dataset
(x_train,y_train),(x_test,y_test) = keras.datasets.mnist.load_data()
len(x_train)
```

```python
fig, axes = plt.subplots(10,10)

k=0
for i in range (10):
    for j in range (10):
        axes[i][j].imshow(x_train[k])
        axes[i][j].axis('off')
        k+=1
        
```

```python
x_train.shape[0]
```

```python
x_train_flatten = x_train.reshape(x_train.shape[0],-1)
index = np.where(y_train == 0)[0]
x_train_new = np.delete(x_train_flatten, index,axis=0)
y_train_new = np.delete(y_train,index)

x_test_flatten = x_test.reshape(x_test.shape[0],-1)
index = np.where(y_test==0)
x_test_new = np.delete(x_test_flatten,index,axis=0)
y_test_new = np.delete(y_test,index)
```

```python
model = Sequential([
    Dense(256,activation='softmax'),
    Dense(128,activation='softmax'),
    Dense(10,activation='softmax'),
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train_new, y_train_new, validation_split=0.2, batch_size=1000, epochs=20)
model.evaluate(x_test_new, y_test_new)
```

```python
y_pred = model.predict(x_test_new)
y_pred[0]
```
