# Model Building

```python
## Fix templets for model testing

def performance(model):
    results = model.evaluate(x_test,y_test)
    print(f"Test Loss {results[0]},\nTest Accuracy {results[1]}")

def grap(history):
    fig, axes = plt.subplots(2,1,figsize=(10,10))
    plt.suptitle('Loss and Accuracy')

    axes[0].set_title('Loss Vs Validation Loss')
    axes[0].plot(history.history['loss'])
    axes[0].plot(history.history['val_loss'])
    axes[0].legend(['Loss', 'Validation loss'])

    axes[1].set_title('Accuracy Vs Validation Accuracy')
    axes[1].plot(history.history['accuracy'])
    axes[1].plot(history.history['val_accuracy'])
    axes[1].legend(['Accuracy','Validation accuracy'])


    plt.show()
    
```

### Model One with only one layer 

```python
modelOne = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(units = 10, activation='sigmoid'),
])
modelOne.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
historyOne = modelOne.fit(x_train,y_train,epochs=20,validation_split=0.2)
```

```python
performance(modelOne)
```

```python
grap(historyOne)
```

```python
from mlxtend.evaluate import bias_variance_decomp


avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        modelOne, x_train, y_train, x_test, y_test, 
        epochs=20, 
        verbose=1)


print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape and normalize image data
x_train = x_train.reshape(len(x_train), 28*28).astype('float32') / 255.0
x_test = x_test.reshape(len(x_test), 28*28).astype('float32') / 255.0
```

```python
# Define baseline model architecture
baseline_model = Sequential()
baseline_model.add(Dense(64, activation='relu', input_shape=(28*28,)))
baseline_model.add(Dense(10, activation='softmax'))
baseline_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(),metrics=['accuracy'])

baseline_history = baseline_model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
# Overfitting Model (Higher Complexity)
overfit_model = Sequential()
overfit_model.add(Dense(256, activation='relu', input_shape=(28*28,)))
overfit_model.add(Dense(128, activation='relu'))
overfit_model.add(Dense(64, activation='relu'))
overfit_model.add(Dense(10, activation='softmax'))
overfit_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(),metrics=['accuracy'])

overfit_history = overfit_model.fit(x_train, y_train, epochs=100, validation_split=0.2)
```

```python
# Underfitting Model (Lower Complexity)
underfit_model = Sequential()
underfit_model.add(Dense(32, activation='relu', input_shape=(28*28,)))
underfit_model.add(Dense(10, activation='softmax'))
underfit_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(),metrics=['accuracy'])

underfit_history = underfit_model.fit(x_train, y_train, epochs=2, validation_split=0.2)
```
