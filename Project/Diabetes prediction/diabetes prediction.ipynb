# Encoding

```python
dataset["gender"].unique()
```

```python
dataset["gender"] = dataset["gender"].map({"Other": 0, "Male": 1,"Female": 2})
dataset.head()
```

# Train_test_split

```python
Y = dataset["diabetes"]
X = dataset.drop("diabetes",axis=1)
```

```python
# 80% 20% split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

# Model selection

```python
from sklearn.tree import DecisionTreeClassifier   #importing algo
from sklearn.metrics import accuracy_score        # for checking accuracy

dtc = DecisionTreeClassifier(criterion='entropy',splitter='best', max_depth=3, max_leaf_nodes=5, random_state=0)
dtc.fit(x_train,y_train)   # traing model


y_pred = dtc.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)


print(f"Accuracy : {accuracy*100 :.2f}%")
```
