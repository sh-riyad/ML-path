#### Unique Value frequency

```python
for col in dataset.columns:
    if col == "Id":
        continue
    else:
        sns.histplot(dataset[col],kde=True)
        plt.show()
```

```python


# Exclude 'Id' column from subplots
columns_to_plot = [col for col in dataset.columns if col != 'Id']

# Set the number of figures per row
figures_per_row = 3

# Calculate the number of rows needed
num_rows = (len(columns_to_plot) + figures_per_row - 1) // figures_per_row

# Create subplots
fig, axes = plt.subplots(num_rows, figures_per_row, figsize=(15, 5 * num_rows))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through columns and create histograms
for i, col in enumerate(columns_to_plot):
    sns.histplot(dataset[col], kde=True, ax=axes[i])
    axes[i].set_title(col)

# Remove any empty subplots
for j in range(len(columns_to_plot), num_rows * figures_per_row):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
